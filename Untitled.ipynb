```python
import pandas as pd
import random
import numpy as np

# Configuration
n_samples = 15000
random.seed(42)
np.random.seed(42)

# Paramètres possibles
sexes = ['Homme', 'Femme']
pays_origine = ['RDC', 'Angola', 'Rwanda', 'Burundi', 'Ouganda', 'Tanzanie', 'Zambie']
raisons_voyage = ['Tourisme', 'Affaires', 'Transit', 'Migration', 'Études']
provenances = ['Kinshasa', 'Lubumbashi', 'Kisangani', 'Goma', 'Kigali', 'Lusaka']
destinations = ['Mbujimayi', 'Kananga', 'Tshikapa', 'Kinshasa', 'Luanda', 'Nairobi']

def generate_data(n):
    data = []
    for _ in range(n):
        age = random.randint(18, 70)
        sexe = random.choice(sexes)
        pays = random.choice(pays_origine)
        raison = random.choice(raisons_voyage)
        provenance = random.choice(provenances)
        destination = random.choice(destinations)
        papiers_valides = random.choice([0, 1])
        casier_judiciaire = random.choice([0, 1])

        # Logique simple pour générer le risque
        score_risque = 0
        if papiers_valides == 0:
            score_risque += 2
        if casier_judiciaire == 1:
            score_risque += 2
        if raison == 'Migration':
            score_risque += 1
        if age < 25 and raison in ['Migration', 'Transit']:
            score_risque += 1

        if score_risque >= 4:
            niveau_risque = 'Élevé'
        elif score_risque >= 2:
            niveau_risque = 'Moyen'
        else:
            niveau_risque = 'Faible'

        data.append([
            age, sexe, pays, raison, provenance, destination,
            papiers_valides, casier_judiciaire, niveau_risque
        ])

    return pd.DataFrame(data, columns=[
        'age', 'sexe', 'pays_origine', 'raison_voyage', 'provenance', 'destination',
        'papiers_valides', 'casier_judiciaire', 'niveau_risque'
    ])

# Génération
df = generate_data(n_samples)

# Sauvegarde
df.to_csv("donnees_migratoires_mbujimayi.csv", index=False)
print("Fichier 'donnees_migratoires_mbujimayi.csv' généré avec succès.")
```
